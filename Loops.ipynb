**While Loop**

```python
i = 1        # Initialize i to 1
n = 5        # Set the value of n to 5

while i <= n:     # Start a while loop. It will run as long as i is less than or equal to n.
    print(i)      # Print the current value of i
    i = i + 1     # Increment i by 1 (i.e., i becomes its current value + 1)
```

```python
n = int(input("Enter number for factorial"))  # Prompt the user to enter a number and convert it to an integer
fact = 1  # Initialize a variable `fact` to 1

if n < 0:  # Check if the entered number is negative
    print("Invalid Factorial: Negative number")  # If negative, print an error message
elif n == 0:  # Check if the entered number is 0
    print("Factorial of 0 = 1")  # If 0, print the factorial as 1 (0! = 1)
else:
    for j in range(1, n + 1):  # Loop from 1 to n (inclusive)
        fact = fact * j  # Calculate the factorial iteratively

    print(f"Factorial of {n} = {fact}")  # Print the calculated factorial
```

```python
i = 1  # Initialize the variable i to 1

while i < 6:  # Start a while loop. It will run as long as i is less than 6.
    print(i)  # Print the current value of i
    if i == 3:  # Check if i is equal to 3
        break  # If i is 3, exit the loop immediately
    i += 1  # Increment i by 1 (i.e., i becomes its current value + 1)
```

```python
x = 0
while True:  # Outer loop runs indefinitely
    print(x)
    if x >= 9:  # Check if x has reached 9
        break   # If x is 9 or greater, exit the outer loop
    x += 1     # Increment x by 1 on each iteration
    print("inside Loop")  # Print "inside Loop" inside the outer loop
```

```python
i = 1  # Initialize i to 1

while i < 10:  # Start a while loop with the condition i < 10
    print(i)  # Print the current value of i
    i = i + 1  # Increment i by 1

else:  # This block is executed when the while loop condition becomes False
    print("End: i is larger than or equal to 10")
```
